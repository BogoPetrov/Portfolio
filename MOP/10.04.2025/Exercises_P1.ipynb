```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Ex. 1
def ex_1():
    print("\x1B[4m" + "Exercise 1:" + "\x1B[0m")

    a = np.array([2, 1])
    b = np.array([-1, 3])
    c = a + b

    plt.quiver(*[0, 0], *a, angles='xy', scale_units='xy', scale=1, color='r', label='a')
    plt.quiver(a[0], a[1], *b, angles='xy', scale_units='xy', scale=1, color='g', label='b')
    plt.quiver(*[0, 0], *c, angles='xy', scale_units='xy', scale=1, color='b', label='c = a + b')
    plt.title('Vector addition in 2D')
    plt.xlim(-1, 5)
    plt.ylim(-1, 5)
    plt.grid()
    plt.legend()
    plt.gca().set_aspect('equal')
    plt.show()
    print("-----" * 20)


# Ex. 2
def ex_2():
    print("\x1B[4m" + "Exercise 2:" + "\x1B[0m")

    u = np.array([3, 0])
    v = np.array([0, -5])
    dot_product = np.dot(u, v)
    print("Vector u:", u)
    print("Vector v:", v)
    print(f"The dot product is {dot_product}. Vectors are orthogonal." if dot_product == 0 else f" The dot product is {dot_product}. Vectors are NOT orthogonal.")
    print("-----" * 20)


def ex_3():
    print("\x1B[4m" + "Exercise 3:" + "\x1B[0m")

    p = np.array([1, 2, 3])
    q = np.array([2, -1, 1])
    cross_product = np.cross(p, q)

    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    origin = np.array([0, 0, 0])

    ax.quiver(*origin, *p, color='r', label='p', arrow_length_ratio=0.35, linewidth=2)
    ax.quiver(*origin, *q, color='g', label='q', arrow_length_ratio=0.35, linewidth=2)
    ax.quiver(*origin, *cross_product, color='b', label='Cross Product p × q', arrow_length_ratio=0.35, linewidth=2)

    ax.set_xlim([-10, 6])
    ax.set_ylim([-2, 6])
    ax.set_zlim([-10, 6])
    ax.set_xlabel('X axis')
    ax.set_ylabel('Y axis')
    ax.set_zlabel('Z axis')
    ax.set_title('3D Visualization')
    ax.legend(loc='upper left', bbox_to_anchor=(1, 1))
    ax.view_init(elev=25, azim=35)
    plt.gca().set_aspect('equal')
    plt.show()

    dot_product_with_p = np.dot(cross_product, p)
    dot_product_with_q = np.dot(cross_product, q)

    print("Cross Product p × q:", cross_product)
    print("Dot Product (p × q) • p:", dot_product_with_p)
    print("Dot Product (p × q) • q:", dot_product_with_q)

    if dot_product_with_p == 0 and dot_product_with_q == 0:
        print("The cross product is perpendicular to both p and q.")
    else:
        print("The cross product is NOT perpendicular to both p and q.")
    print("-----" * 20)


# Ex. 4
def ex_4():
    print("\x1B[4m" + "Exercise 4:" + "\x1B[0m")

    a = np.array([4, 2])
    b = np.array([1, 5])
    c = a - b

    plt.quiver(*[0, 0], *a, angles='xy', scale_units='xy', scale=1, color='r', label='a')
    plt.quiver(*[0, 0], *b, angles='xy', scale_units='xy', scale=1, color='g', label='b')
    plt.quiver(b[0], b[1], *c, angles='xy', scale_units='xy', scale=1, color='b', label='c = a - b')
    plt.title('Vector subtraction in 2D')
    plt.xlim(-1, 8)
    plt.ylim(-1, 8)
    plt.legend()
    plt.grid()
    plt.gca().set_aspect('equal')
    plt.show()
    print("-----" * 20)


# Ex. 5
def ex_5():
    print("\x1B[4m" + "Exercise 5:" + "\x1B[0m")

    r = np.array([3, -4, 12])
    norm_r = np.linalg.norm(r)
    normalized_r = r / norm_r
    norm_normalized_r = np.linalg.norm(normalized_r)

    print(f"Vector r: {r}")
    print(f"Length of r: {norm_r}")
    print(f"Normalized vector: {normalized_r}")
    print(f"Length of normalized vector: {norm_normalized_r}")
    print("-----" * 20)

# Ex. 6
def ex_6():
    print("\x1B[4m" + "Exercise 6:" + "\x1B[0m")

    a = np.array([1, 1])
    b = np.array([2, 0])
    angle_between = np.rad2deg(np.arccos(np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))))
    print(f"Vector a: {a}")
    print(f"Vector b: {b}")
    print(f"Angle between vectors a and b: {angle_between:.2f}°")
    print("-----" * 20)

# Ex. 7
def ex_7():
    print("\x1B[4m" + "Exercise 7:" + "\x1B[0m")

    u = np.array([3, 4])
    v = np.array([2, 1])
    projection_u_on_v = (np.dot(u, v) / np.linalg.norm(v) * np.linalg.norm(v)) * v
    projection_u_on_v = np.int32(projection_u_on_v)

    print(f"Vector u: {u}")
    print(f"Vector v: {v}")
    print(f"Projection of v onto u: {projection_u_on_v}")
    print("-----" * 20)

# Ex. 8
def ex_8():
    print("\x1B[4m" + "Exercise 8:" + "\x1B[0m")

    point_a_coordinates = np.array([1, 0, 2])
    d = np.array([2, -2, -1])
    t_values = np.linspace(-2, 2, 20)

    points = np.array([point_a_coordinates + t * d for t in t_values])

    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot(points[:, 0], points[:, 1], points[:, 2], label='r(t) = A + t·d')
    ax.scatter(*point_a_coordinates, color='red', label='Point A', s=50)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title('Parametric line in 3D')
    ax.legend()
    plt.gca().set_aspect('equal')
    plt.tight_layout()
    plt.show()
    print("-----" * 20)


# Ex. 9
def ex_9():
    print("\x1B[4m" + "Exercise 9:" + "\x1B[0m")

    u = np.array([1, 2])
    v = np.array([1, 1])
    a_values = np.linspace(-2, 2, 5)
    b_values = np.linspace(-2, 2, 5)

    combinations = []
    for a in a_values:
        for b in b_values:
            combinations.append(a * u + b * v)

    combinations = np.array(combinations)

    plt.axhline(0, color='gray', linewidth=0.5)
    plt.axvline(0, color='gray', linewidth=0.5)
    plt.grid(True, linestyle='--', alpha=0.5)

    plt.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='blue', label='u')
    plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='green', label='v')

    plt.scatter(combinations[:, 0], combinations[:, 1], color='red', label='Linear combinations', s=30)

    plt.legend()
    plt.title('Linear combinations of u and v')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.axis('equal')
    plt.tight_layout()
    plt.show()
    print("-----" * 20)

# Ex. 10
def ex_10():
    print("\x1B[4m" + "Exercise 10:" + "\x1B[0m")

    a = np.array([3, 1])

    rotated_45_deg = rotate_2d(a, 45)
    print(f"Original vector: {a}")
    print(f"Rotated by 45°: {rotated_45_deg}")

    rotated_90_deg = rotate_2d(a, 90)
    print(f"Rotated by 90°: {rotated_90_deg}")

    plt.quiver(a[0], a[1], *a, angles='xy', scale_units='xy', scale=1, color='blue', label='a')
    plt.quiver(rotated_45_deg[0], rotated_45_deg[1], *rotated_45_deg, angles='xy', scale_units='xy', scale=1, color='red', label='a rotated by 45°')
    plt.quiver(rotated_90_deg[0], rotated_90_deg[1], *rotated_90_deg, angles='xy', scale_units='xy', scale=1, color='green', label='a rotated by 90°')
    plt.title('Rotation of vector a in 2D')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.xlim(-10, 10)
    plt.ylim(-10, 10)
    plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
    plt.grid()
    plt.gca().set_aspect('equal')
    plt.show()

def rotate_2d(vector, angle_deg):
    rad = np.deg2rad(angle_deg)
    r = np.array([
        [np.cos(rad), -np.sin(rad)],
        [np.sin(rad), np.cos(rad)]
    ])
    return r @ vector
```

```python
# Run exercises form 1 to 10
ex_1()
ex_2()
ex_3()
ex_4()
ex_5()
ex_6()
ex_7()
ex_8()
ex_9()
ex_10()
```
