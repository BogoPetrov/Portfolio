```python
import numpy as np

v = np.array([3, 4])
length = np.linalg.norm(v)
print(f"Дължината на вектора е: {length}")
```

```python
a = np.array([1, 2])
b = np.array([-2, 1])

dot_product = np.dot(a, b)

if dot_product == 0:
    print("Векторите са ортогонални.")
else:
    print("Векторите не са ортогонални.")
```

```python
import matplotlib.pyplot as plt

# Вектори
a = np.array([3, 2])
b = np.array([-1, 4])

# Начална точка (0,0)
origin = np.array([0, 0])

# Плот на векторите
plt.quiver(*origin, *a, color='r', angles='xy', scale_units='xy', scale=1, label='a')
plt.quiver(*origin, *b, color='b', angles='xy', scale_units='xy', scale=1, label='b')

# Задаване на граници и мрежа
plt.xlim(-2, 5)
plt.ylim(-2, 5)
plt.axhline(y=0, color='black', linewidth=0.5)
plt.axvline(x=0, color='black', linewidth=0.5)
plt.grid()
plt.legend()
plt.title("Графично представяне на вектори")

# Показване на графиката
plt.show()
```

```python
a = np.array([1, 0, 2])
b = np.array([2, 1, -1])

cross_product = np.cross(a, b)
print(f"Векторното произведение е: {cross_product}")
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Вектори
a = np.array([1, 2, 3])
b = np.array([4, -1, 2])

# Начална точка (0,0,0)
origin = np.zeros(3)

# Плот на векторите
ax.quiver(*origin, *a, color='r', label='a')
ax.quiver(*origin, *b, color='b', label='b')

# Оси
ax.set_xlim([0, 5])
ax.set_ylim([-2, 5])
ax.set_zlim([0, 5])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title("3D представяне на вектори")
ax.legend()

plt.show()
```
