```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Създаване на фигура и 3D оси
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Данни за линии
x = np.linspace(-5, 5, 100)
y = np.sin(x)
z = np.cos(x)

# 1. Чертаене на 3D линия
ax.plot(x, y, z, color='b', label='3D Линия')

# 2. Чертаене на точки
x_points = np.random.uniform(-5, 5, 10)
y_points = np.random.uniform(-5, 5, 10)
z_points = np.random.uniform(-5, 5, 10)
ax.scatter(x_points, y_points, z_points, color='r', marker='o', label='Точки')

# 3. Чертаене на вектор
ax.quiver(0, 0, 0, 2, 3, 1, color='g', label='Вектор')

# 4. Чертаене на повърхност
X, Y = np.meshgrid(np.linspace(-5, 5, 30), np.linspace(-5, 5, 30))
Z = np.sin(np.sqrt(X**2 + Y**2))
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.5)

# Настройки на визуализацията
ax.set_xlabel('X ос')
ax.set_ylabel('Y ос')
ax.set_zlabel('Z ос')
ax.set_xlim([-5, 5])
ax.set_ylim([-5, 5])
ax.set_zlim([-5, 5])
ax.view_init(elev=30, azim=45)  # Изглед под ъгъл

ax.legend()
plt.title("Пример с Axes3D в Matplotlib")
plt.show()
```
