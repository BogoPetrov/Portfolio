```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Ex. 1
a = np.array([3, 2])
b = np.array([-1, 4])
c = a + b

plt.figure()
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='r', label='a')
plt.quiver(a[0], a[1], b[0], b[1], angles='xy', scale_units='xy', scale=1, color='b', label='b')
plt.quiver(0, 0, c[0], c[1], angles='xy', scale_units='xy', scale=1, color='g', label='c = a + b')
plt.xlim(-2, 8)
plt.ylim(-2, 8)
plt.grid()
plt.legend()
plt.title("Събиране на вектори")
plt.show()
```

```python
# Ex. 2
u = np.array([4, -2])
v = np.array([1, 2])

dot_product = np.dot(u, v)

if dot_product == 0:
    print("Векторите са ортогонални.")
else:
    print("Векторите не са ортогонални.")

print(f"Скаларно произведение: {dot_product}")
```

```python
# Ex. 3
p = np.array([1, 2, 3])
q = np.array([4, 0, -1])
r = np.cross(p, q)

origin = np.zeros(3)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.quiver(*origin, *p, color='r', label='p')
ax.quiver(*origin, *q, color='b', label='q')
ax.quiver(*origin, *r, color='g', label='r = p × q')

ax.set_xlim([-5, 5])
ax.set_ylim([-5, 5])
ax.set_zlim([-5, 5])

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.legend(bbox_to_anchor=(1, 1), loc=2, borderaxespad=0.5)
ax.set_title("Векторно произведение в 3D")
plt.show()

print(f"Векторно произведение: {r}")
```
